```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset 
import torch

# change me!
modelpath="models/TinyLlama-1.1B-Chat-v1.0"

model = AutoModelForCausalLM.from_pretrained(
    modelpath,    
    torch_dtype=torch.bfloat16,
    device_map="auto",
    attn_implementation="flash_attention_2",
)
tokenizer = AutoTokenizer.from_pretrained(modelpath) 
```

# Prompts from list of strings

```python
from utils import ModelPredictionGenerator

questions = [
    "Hey who are you?",
    "say something",
    "say nothing",
    "whats your favourite color?",
    "why is x=2?"
]

gen = ModelPredictionGenerator(model, tokenizer)

predictions = gen.run(input_data = questions, max_new_tokens = 100)

for i, p in enumerate(predictions):
    print("#"* 10, f"Prediction #{i} ({p['tok/s']} tok/s)")
    print(p["answer_pred"], "\n")    
```
